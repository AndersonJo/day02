# 한화짱 

우리는 정말 훌륭한 회사에서 일하고 있다

- **볼드체**
- ~~스트라이크~~

$$ x^2 + \frac{5}{x}  $$

```python
def classify(w):
    if pd.isna(w): return 'Unknown'
    if w >= 1000: return 'Heavy'
    else: return 'Light'        
products.product_weight_g.apply(classify)
```
