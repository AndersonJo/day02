# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv('./data/olist_customers_dataset.csv')
orders = pd.read_csv('./data/olist_orders_dataset.csv')
items = pd.read_csv('./data/olist_order_items_dataset.csv')
payments = pd.read_csv('./data/olist_order_payments_dataset.csv')
reviews = pd.read_csv('./data/olist_order_reviews_dataset.csv')
products = pd.read_csv('./data/olist_products_dataset.csv')
sellers = pd.read_csv('./data/olist_sellers_dataset.csv')
transaction = pd.read_csv('./data/product_category_name_translation.csv')
```

```python
step1 = pd.merge(orders, customers, on='customer_id', how='inner')
step2 = pd.merge(step1, items, on='order_id', how='inner')
step3 = pd.merge(step2, products, on='product_id', how='inner')
step4 = pd.merge(step3, sellers, on='seller_id', how='inner')
master = pd.merge(step4, transaction, on='product_category_name', how='inner')
```

```python
master['total_paid'] = master['price'] + master['freight_value']
master['order_purchase_timestamp'] = pd.to_datetime(master['order_purchase_timestamp'])
```

```python
master.groupby('year_month')['total_paid'].sum()
```
